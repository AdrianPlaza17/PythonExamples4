Visualización de datos

Utilizando una fuente de datos, consumirla en Python para mostrar diferentes tipos de
gráficos


La fuente de datos que he elegido es sobre jugadores de futbol de ligas europeas.

```python
import pandas as pd

df=pd.read_csv("Data.csv")

print (df.head(10))
```

### 3.1.1. Gráfico de barras: Goles de cada jugador

```python
import matplotlib.pyplot as plt
import numpy as np


x = df["Player Names"].head(20)
y = df["Goals"].head(20)


# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)



plt.xticks(rotation='vertical')

plt.show()
```

### 3.1.2. Gráfico circular: Goles por partido de cada jugador

```python
import matplotlib.pyplot as plt

golesPartido=df["Goals"]/df["Matches_Played"]

labels = df["Player Names"].head(10)

sizes = golesPartido.head(10)


fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, shadow=True, startangle = 90)


plt.show()
```

### 3.1.3. Gráfico combinado: Grafico de tiros y tiros a puerta

```python




data1 = df["Shots"].head(10)
data2 = df["OnTarget"].head(10)

fig, ax1 = plt.subplots()

color = 'tab:red'

ax1.plot(data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:blue'

ax2.plot(data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()
plt.show()
```
